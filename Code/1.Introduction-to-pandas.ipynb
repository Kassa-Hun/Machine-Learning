```python
# Adding Column from Python list
fuel_economy=[7.5,9.2,5.0,9.6,8.7,6.4,9.0,4.5,7.5,6.9]
carSales["Fuel per 100KM"]=fuel_economy
carSales
```

```python
# Creating a column from an existing column
# How many litters a car used in its entire life?
carSales["Total fuel used (L)"]=carSales["Odometer (KM)"] /100 * carSales["Fuel per 100KM"]
carSales
```

```python
# Create a column from a single value
carSales["Number of wheels"]=4
carSales["Passed road safety"]=True
carSales
```

```python
# Deleting a column
carSales.drop("Total fuel used (L)",axis=1,inplace=True)
```
